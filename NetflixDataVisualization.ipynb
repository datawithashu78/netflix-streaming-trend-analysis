# Data Visualization

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('clean_netflix_data.csv')
```

```python
sns.set_style('whitegrid')
```

# What is the most frequent genre of movies released on Netflix

```python
df['Genre'].describe()
```

Ans: The most frequent genre of movies released on Netflix is Drama.

```python
sns.catplot(y = 'Genre' ,
            data = df,
            kind ='count',
            order=df['Genre'].value_counts().index,
            color='#4287f5')
plt.title('Genre column distribution')
plt.show()
```

# Which has highest votes in vote average column

```python
df.head()
```

```python
plot = sns.catplot(
    y='Vote_Average',
    data=df,
    kind='count',
    order=df['Vote_Average'].value_counts().index,
    height=6,
    aspect=1.4,
    color='Green'
),
plt.title('Votes distribution')
plt.show()
```
